# Data Extraction

```python
# Explanation: This cell contains code that starts with: `df.columns`
df.columns
```

```python
# Explanation: This cell contains code that starts with: `data = df[['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol']]`
data = df[['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol']]
```

```python
# Explanation: Prepares data for training and fits a machine learning model.
train, test = train_test_split(data, test_size=0.25, random_state=1)
#Split dataframe into training and testing sets 
#25% of data goes to testing split, 75% of data goes to training split, random_state=1 ensures reproducible results
```

```python
# Explanation: This cell contains code that starts with: `actual = np.concatenate((np.ones(45), np.zeros(450), np.ones(5)))`
actual = np.concatenate((np.ones(45), np.zeros(450), np.ones(5)))
actual
# Create a synthetic numpy array representing 'actual' class labels for demonstration.
# It combines 45 ones, followed by 450 zeros, followed by 5 ones.
```

```python
# Explanation: This cell contains code that starts with: `predicted = np.concatenate((np.ones(100), np.zeros(400)))`
predicted = np.concatenate((np.ones(100), np.zeros(400)))
predicted
# Create a synthetic numpy array representing 'predicted' class labels for demonstration.
# It combines 100 ones followed by 400 zeros.
```
